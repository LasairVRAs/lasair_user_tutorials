## What you'll learn

- How is the `absMag` feature in Lasair calculated. 

### Background

The absolute magnitude of an object is the apparent magnitude it would have if located 10 parsec away from us. 
There are three factors that come into play:
* **Distance** - the obvious one. We need to correct for how far away the object is (it's distance modulus)
* **Extinction** - if there is dust in the way we need to correct for the fact that the apparent magnitude we measure is fainter due to extinction
* **Redshift** - This is called a K-correction. In Lasair, an approximate version of the k-correction is performed (see [their manual](https://lasair-lsst.readthedocs.io/en/main/cookbook.html#absolute-magnitude))

In Lasair the absolute magnitude is calculated with all 3 corrections, as of 2026-03-03 if there is not redshift but direct distance is known, the `absMag` feature will still be null. This is being fixed.

<h2 align="center">  1. Distance, Redshift and E(B-V) for an example </h2>

Here we are using the API to get some values for an object I have selected. If you want to know how the API works go to the `tutorials/API_recipes.ipynb`

````python
from lasair import lasair_client as lapi
import os
import numpy as np

API_TOKEN = os.getenv('LASAIR_LSST_TOKEN')

if API_TOKEN is None:
    print("No Token found. Check Spelling. Note that if you have just added your token to your environment variables, you may need to restart your terminal so your shell settings are reloaded. ")
````

````python
# first let's get a pair and ebv from a known source
L = lapi(API_TOKEN, 
         endpoint='https://api.lasair.lsst.ac.uk/api'
         )
````

````python
result = L.object(170019716333043819, 
                  lasair_added=True)
````

### Redshift from Sherlock:

When you get the data back from the API it is nested in this `lasairData` key where we can find the `sherlock` dictionary amongst other things. 

This does not follow the [schema](https://lasair.lsst.ac.uk/schema/) we use to make filters. I am not sure why but if you are looking for sherlock in your API calls know that it is here: 

````python
result['lasairData']['sherlock']
````

````python
import astropy.cosmology.units as cu
z = result['lasairData']['sherlock']['photoZ'] * cu.redshift
````

````python
from astropy.cosmology import WMAP9
import astropy.units as u

d = z.to(u.pc, cu.redshift_distance(WMAP9, kind="comoving"))
d
````

### Now let's grab Lasair's value for E(B-V)

_How Did you get that E(B-V)?_

We use the [Schlafly and Finkbeiner ](https://iopscience.iop.org/article/10.1088/0004-637X/737/2/103) corrections. If you have ever used the `dustmaps` python package, that is the `SFD` object, 

In this example I don't make us run `dustmaps` because the dependency comes with heavy dust extinction maps (a heavy dependecy for a lesson). If you already have the dustmaps package feel free to run it to double check! You can get the ra and dec using:

```python
result['diaObject']['ra'], result['diaObject']['decl']

````python
ebv = result['lasairData']['ebv']
ebv
````

<h2 align="center"> The Corrections </h2>

### Distance Modulus

````python
import numpy as np

mu = 5*np.log10(d.value)-5
print(f"Distance Modulus: {mu:.2f}")
````

### Latest Aparent Mag to raw abs mag

````python
# Picking a "bluer" filter so extinction correction is greater for the example ;)
result['diaSourcesList'][2]
````

````python
def nJy2mag(nJy):
    """Convert nJy to AB magnitude
    """
    return -2.5 * np.log10(nJy / 3631e9)
````

````python
raw_absMag = nJy2mag(result['diaSourcesList'][0]['psfFlux']) - mu
print(f"Raw Absolute Magnitude: {raw_absMag:.2f}")
````

### Extinction Correction: Reddening and E(B-V)
We have E(B-V); we need our $R_V$ factor, which is **filter dependent**. For a standard $R_V$ factor of 3.1, we have the following **reddening  factors** (see Table 6 - [Schlafly and Finkbeiner 2011](https://iopscience.iop.org/article/10.1088/0004-637X/737/2/103)

````python
Rv_3p1_dict = {'u': 4.145, 'g': 3.237, 'r': 2.273, 'i': 1.684, 'z': 1.323, 'y': 1.088}
print(f"Reddening in our  {result['diaSourcesList'][2]['band']} band: {Rv_3p1_dict[result['diaSourcesList'][2]['band']]:.2f}")
````

````python
print(f"The Extinction is A_i = {Rv_3p1_dict[result['diaSourcesList'][2]['band']]*ebv:.2f} magnitudes")
````

````python
extinction_corrected_absMag = raw_absMag - Rv_3p1_dict[result['diaSourcesList'][2]['band']]*ebv
print(f"Extinction Corrected Absolute Magnitude: {extinction_corrected_absMag:.2f}")
````

### K-Correction

We are using [Hogg et al. 2002 (eq. 7 but turned into mags)](https://arxiv.org/pdf/astro-ph/0210394)

````python
k_correction =  2.5*np.log10(1+z)
print(f"K-correction: {k_correction:.2f} magnitudes")
````

````python
final_absMag = extinction_corrected_absMag + k_correction
print(f"Final Absolute Magnitude: {final_absMag:.2f}")
````
